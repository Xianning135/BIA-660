```python
import requests as r
import re
from bs4 import BeautifulSoup
import csv
```

```python
tld=r.get("https://en.m.wikipedia.org/wiki/List_of_Internet_top-level_domains")
```

```python
soup = BeautifulSoup(tld.content, 'html.parser')
```

```python
class_lists=soup.find_all(class_='wikitable sortable')
```

```python
list1=[]
for class_list in class_lists:
    items=class_list.find_all(valign='top')
    list1.append(items)
```

```python
list2=[]
for i in list1:
    for j in i:
        name=j.contents[0]
        list2.append(name)
```

```python
list3=[]
for i in list2:
    a=i.find_all('a')
    list3.append(a)
```

```python
list4=[]
for i in list3:
    for j in i:
        href=j.get('href')
        list4.append(href)
```

```python
list4[9]
```

```python
list5=[]
for i in list4:
    if i[3]=='k':
        list5.append(i)
```

```python
list5
```

```python
list6=[]
for i in list5:
    domain=i.strip('/wiki/')
    list6.append(domain)
```

```python
list6
```

```python
sld='.example'
urls=[]
for domain in list6:
    url='http://www'+sld+domain
    urls.append(url)
```

```python
urls
```

```python
check_lists=[]
for i in urls:
    try:
        check = r.get(i)
        status=check.status_code
        check_lists.append(status)
    except:
        error=i + ' is not valid'
        check_lists.append(error)
```

```python
check_lists
```

```python
with open('results.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerows(zip(list6, check_lists))
```
