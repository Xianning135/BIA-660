```python
sentten=[]
for i in range(len(sentences)):
    if len(sentences[i]) > 10:
        sentten.append(sentences[i])
```

```python
similarity=[]
sent1=[]
sent2=[]
for i in range(len(sentten)):
    for j in range(len(sentten)):
        sent1.append(i)
        sent2.append(j)
        similarity.append(sentten[i].similarity(sentten[j]))
```

```python
import pandas as pd
df=pd.DataFrame(
    {'sentence 1': sent1,
     'sentence 2': sent2,
     'similarity': similarity
    })
```
